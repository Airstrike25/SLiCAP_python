```python
A, B, C, D, V_i, I_i, V_o, I_o = sp.symbols('A B C D V_i, I_i, V_o, I_o')
T1 = sp.Matrix([[A, B], [C, D]])
in_ = sp.Matrix([[V_i],[I_i]])
out = sp.Matrix([[V_o],[I_o]])
HTML('$$' + sp.latex(in_) + '=' + sp.latex(T1) + '\cdot' + sp.latex(out) + '$$')
```

```python
HTML(head2html('Source-to-load transfer'))
```

```python
instr.setSimType('symbolic')
instr.setGainType('gain')
instr.setDataType('laplace')
instr.setSource('I1')
instr.setDetector('V_out')
s_l_transfer = sp.simplify(instr.execute().laplace)
HTML('The source-to-load transfer can be written as:')
```

```python
HTML(eqn2html('V_out/I_s', s_l_transfer))
```

```python
HTML(text2html("The transimpedance gain does not depend on the source and the load impedance if A=B=D=0:"))
```

```python
s_l_transfer_transimpedance = s_l_transfer.subs({A:0, B:0, D:0})
HTML(eqn2html('V_out/I_s', s_l_transfer_transimpedance))
```
